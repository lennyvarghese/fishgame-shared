Two "response" coding types: "stimulus" coding (response column corresponds to what user actually pressed) and "accuracy" coding (response column corresponds to whether user got trial correct or not). Used 10000 sample iterations, burn 2000. For the purpose of getting quick estimates, did not include inter-trial variability parameters.

## "Response coding"

Assume v and z depend on congruence. The "response" variable here is whatever the subject answered (fast or slow). The stimulus column is marked (fast or slow). 

Put a link function on v to flip the sign depending on the stimulus. The link function on z is just the standard 1/(1+exp(-x)).

### Visual

```python
m = hddm.load('visual_quick_response_v_with_flip_z/model_1.pkl')
```

```python
m.plot_posteriors('v_Intercept')
m.plot_posteriors('v_C(condition, Treatment("control"))[T.congruent]')
m.plot_posteriors('v_C(condition, Treatment("control"))[T.incongruent]')
```

## Accuracy coding

Same as above, but using "accuracy coding" instead (response encodes correct or incorrect). This analysis is similar to that shown on the HDDM help site for "stimulus coding with HDDMRegression".

Here, instead of "v" being flipped, "z" gets flipped based on the stimulus type and then run through the logistic function.
